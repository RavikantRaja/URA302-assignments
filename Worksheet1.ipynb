```python
year = int(input("Enter the year"))
if year%100 == 0:
  if year%400 == 0:
    print("Entered year is a leap year")
  else:
    print("Entered year is not a leap year")
else:
  if year%4 == 0:
    print("Entered year is a leap year")
  else:
    print("Entered year is not a leap year")
```

```python
x1 = int(input("Enter the x coordinate of 1st point "))
y1 = int(input("Enter the y coordinate of 1st point "))
x2 = int(input("Enter the x coordinate of 2nd point "))
y2 = int(input("Enter the y coordinate of 2nd point "))
cord = [x2-x1,y2-y1]
distance = math.sqrt((cord[0]**2+cord[1]**2))
print("Distance between the two points is",distance)
```

```python
num = int(input("Enter a number "))
fac = 0
for i in range(2,int(math.sqrt(num)+1)):
  if num%i==0:
    fac = 1
if fac == 1:
  print("Entered number is not a prime number")
else:
  print("Entered number is a prime number")
```

```python
n = int(input("Enter a number "))
ans = 0
for i in range(1,n+1):
  ans = ans+i**2
print("Sum of square of first",n,"natural numbers is",ans)
```
